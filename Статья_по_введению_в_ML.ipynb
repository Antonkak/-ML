## Что вообще за зверь "Машинное обучение"

Давайте разберёмся, что называется машинным обучением. Например, мы хотим определять, кто представлен на изображении: собака или кошка. Для решения этой задачи мы написали программу и запустили её на компьютере — той самой «машине». Программа получает новые примеры того, как могут выглядеть кошки и собаки, и учится отличать кошек от собак на новых изображениях. Чем больше обучающих данных мы предоставим программе, тем больше изображений из тестовой выборки она научится классифицировать правильно.

Машинное обучение — это область, которая создаёт подобные алгоритмы: они не запрограммированны на решение какой-то конкретной задачи (сначала наш алгоритм ничего не знал о кошках и собаках), алгоритмы машинного обучения учатся решать её на основе предоставляемых примеров.

Есть и более формальное определение алгоритма машинного обучения, которое было дано Томом Митчеллом в конце прошлого века:

«Компьютерная программа учится на основе опыта E по отношению к классу задач T и способу оценки качества P, если качество её работы на задачах из T, измеренное с помощью P, растёт с получением опыта E».

Несмотря на формальность, определение использует ровно те концепции, о которых мы говорили в предыдущем примере.

«Программа» в определении Митчелла — тот самый алгоритм, который предназначен для решения поставленной задачи и запущен на нашей машине. Более правильно называть такой алгоритм моделью машинного обучения.

Класс задач, решение которых предполагается с помощью данной модели, действительно задан однозначно — это классификация изображений с кошками и собаками.

«Опыт» из определения — это данные, с помощью которых мы обучали модель. Получение нового опыта — дообучение модели на новых данных.

Способ оценки качества модели тоже был. Мы считали, что модель тем лучше решает задачу, чем выше доля правильных предсказаний, которые она выдаёт на определённом наборе примеров.

Всё получается очень логично: программа называется моделью машинного обучения, если при получении новых данных сама учится решать какую-то задачу всё лучше и лучше.

## Что умеет Машинное обучение

Стоит также сказать, что умеет машинное обучение.

Беспилотные автомобили

Первый пример — беспилотные автомобили. Это автономное транспортное средство, которым управляет не человеком, а искусственным интеллектом. Последние успехи в этой области связаны с применением глубокого обучения, однако ещё в начале XXI века для управления беспилотным транспортом использовали классические алгоритмы машинного обучения.

Допустим, есть машина, и мы поставили на неё камеру, которая в каждый момент времени будет фиксировать состояние дороги. Также мы установили на машину датчики, которые в каждый момент времени будут считывать, что происходит с ней: например, какая педаль нажата и насколько сильно.

После этого мы посадим за руль человека и попросим его проехать по тестовому треку. В процессе поездки мы в каждый момент времени будем снимать показания с камеры и датчиков, формируя обучающую выборку.

В итоге у нас получится размеченная выборка, в которой для каждой конкретной ситуации на дороге представлено действие, которое нужно совершить: например, если на изображении дорога поворачивает направо, то и руль нужно повернуть в ту же сторону.

На основе такой выборки создаём модель машинного обучения, которая учится по состоянию дороги предсказывать, как нужно изменить состояние машины. Теперь мы можем заменить человека за рулём машины обученной моделью и посмотреть, как она будет вести себя на треке.

Эта задача — классический пример задачи обучения с учителем. У нас была размеченная обучающая выборка, в которой каждому состоянию дороги сопоставлено состояние машины. И мы хотели для каждого нового состояния дороги научиться предсказывать, как нужно изменить состояние машины, чтобы продолжить движение.

Генерация изображений по текстовому описанию

Ещё один пример использования технологий машинного обучения — генерация изображений по текстовому описанию. Можно сформулировать абстрактную мысль в виде текста и попросить машину сгенерировать изображение, подходящее под данное описание.

Для создания такой модели нужна большая база неразмеченных изображений. Превратим каждое изображение в некоторый «шум» и попытаемся из него восстановить исходное изображение. Добиваться этого мы будем с помощью двух моделей машинного обучения: кодировщика и декодировщика. Кодировщик превращает исходное изображение в шум, а декодировщик превращает шум в исходник. Если вдруг не получилось восстановить исходное изображение с помощью декодировщика и шума, то достаточно поменять параметры декодировщика так, чтобы получившееся изображение больше походило на исходное. Прогоним таким образом все изображения через кодировщик и декодировщик, чтобы наша модель научилась преобразовывать любое изображение в шум, а потом обратно.

Вспомним, что изначально мы хотели работать с текстами. По аналогичной схеме создадим кодировщик, который умеет превращать текст в шум, а также декодировщик, который умеет восстанавливать исходный текст из шума.

Теперь при запросе на генерацию изображения по текстовому описанию будем делать следующее: сначала превратим текст в шум с помощью кодировщика для текстов, а потом восстановим из полученного шума изображение с помощью декодировщика для изображений. Данное изображение будет абсолютно новым: декодировщик попытался восстановить изображение из шума, для которого не было картинки-прообраза.

Эта задача — пример задачи обучения без учителя, поскольку мы работали с неразмеченными обучающими выборками изображений и текстов. В результате работы наших моделей машинного обучения ожидалось то же самое, что и подавалось им на вход.

Генерация кода по текстовому описанию

Рассмотрим ещё один интересный пример технологий искусственного интеллекта. Уже сейчас существуют модели, которые по текстовому описанию генерируют код для решения задачи, сформулированной ниже.

Допустим, есть заготовка для функции, которая аннотирована следующим комментарием: функция должна реализовывать вычисление n
-го числа Фибоначчи. По такому описанию специальная модель сгенерирует нам тело функции, и полученная функция действительно будет решать поставленную задачу.

В основе моделей, которые занимаются генерацией кода по текстовому описанию, лежат модели, которые занимаются генерацией текста по текстовому описанию. Понятно, что текст на языке программирования — тоже текст, просто на специфическом языке.

Модель для генерации текста на естественном языке по описанию работает следующим образом: она учится по нескольким словам из предложения предсказывать следующее слово в нём. Как можно использовать эту модель, чтобы генерировать абсолютно новый текст? Передадим начало какого-то текста на вход модели — она выдаст нам следующее слово. Далее это новое слово присоединяется к текстовой «затравке», удлиняя её. Новую удлинённую «затравку» снова передадим на вход модели — она выдаст следующее слово, тем самым продолжая текст. Так будем повторять до тех пор, пока модель не закончит генерировать текст.

Вернёмся к изначальной задаче. Текст на языке программирования (код) — это текст на естественном языке плюс текст на данном языке программирования. Здесь естественный язык — то, что сформулировано в комментарии к функции, а текст на языке программирования — это и есть код нужной программы. Обучим новую модель машинного обучения, которая будет уметь работать по описанному выше принципу не только с естественным языком, но и с языками программирования. Для генерации программы по описанию будем передавать модели на вход текстовую затравку в виде комментариев и прокручивать генерацию новых слов (которые в данном случае являются словами на языке программирования), пока модель не остановит работу, и получим в результате программу, которая является продолжением текстового комментария.

В основе подобных моделей для генерации кода по текстовому описанию лежат заранее обученные на большом корпусе текстов модели для генерации текста по текстовому описанию, которые специально дообучаются (это официальный термин), чтобы уметь генерировать в том числе код.

# От слов к началу реализации

Обсудим этапы реализации задачи машинного обучения на примере следующей задачи: на основании данных о том, как ученик учился в полугодии, нужно предсказать его итоговую оценку.


Для начала стоит сказать, что впринципе существует две задачи Машинного обучения: регрессия и классификация. В регрессии мы хотим предсказать конкретное число, например прибыль магазина. В задаче классификации мы ка бы хотим отнести обьект к определенному классу, например нам даны животные нужно из них выделить только котов.




Разберемся с нашей задачей. Хочется предсказать характеристику (итоговую оценку) по имеющимся данным. Чтобы выбрать метод решения задачи, надо определиться с конкретным типом задачи. На первый взгляд кажется, что это задача классификации, так как итоговая оценка может принимать конечное число значений (целое число от 1
 до 5
). Однако эти классы упорядочены: например, оценка «3
» хуже оценки «4
», оценка «4
» хуже оценки «5
» и так далее. Поэтому корректнее рассмотреть эту задачу как задачу регрессии: методы решения задачи регрессии умеют учитывать порядок предсказываемых значений.

Решение задачи машинного обучения делится на несколько этапов:

Получение данных. Нужно собрать данные, на которых мы впоследствии будем обучать модель.
Исследовательский анализ данных. Данные могут содержать неточности и шумы, поэтому прежде чем переходить к созданию модели машинного обучения, нужно специальным образом подготовить эти данные. На этом этапе данные исследуются на предмет скрытых закономерностей.
Подготовка факторов. В машинном обучении характеристики объекта, которые используются для его описания, принято называть факторами. Не все факторы могут быть полезны при построении модели машинного обучения, на этом этапе отбирают только нужные.
Создание модели МО. Следует выбрать модель, которая лучше всего подходит для решения поставленной задачи, а затем обучить её на какой-то части обучающей выборки и проверить, насколько хорошо модель ведёт себя на другой части обучающей выборки. Если её поведение неоптимально, то можно поменять какие-то её параметры и повторить процесс обучения.
Оценка качества. Нужно оценить итоговое качество работы модели на тестовой выборке — данных, которые модель ещё не «видела».

Но чтобы все это реализовать нужно ползнакомится с дополнительными инструментами, библиотеками.

```python
# нарисуем теперь гистограмму для набора случайных величин
import numpy as np

mu = 0
sigma = 10
values = np.random.normal(mu, sigma, 10000)

# the histogram of the data
plt.hist(values, bins=50, density=True)

# интересный факт: для нормального распределения примерно
# 95% выборки будут лежать между mu - 2*sigma и mu + 2*sigma
plt.axvline(x=(mu + 2 * sigma), c='g')
plt.axvline(x=(mu - 2 * sigma), c='r')

# интересный факт: для нормального распределения примерно
# 68% выборки будут лежать между mu - sigma и mu + sigma
plt.axvline(x=(mu + sigma), c='r')
plt.axvline(x=(mu - sigma), c='#ffb899')

plt.xlabel('Value')
plt.ylabel('Estimated probability')

plt.title('Histogram')
plt.grid()

plt.show()
```

## **Метод k ближайших соседей (kNN)**, реализация в библиотеке sklearn.

https://ru.wikipedia.org/wiki/Метод_k-ближайших_соседей



В случае использования метода для классификации объект присваивается тому классу, который является наиболее распространённым среди $k$ соседей данного элемента, классы которых уже известны. В случае использования метода для регрессии, объекту присваивается среднее значение по $k$ ближайшим к нему объектам, значения которых уже известны.

**Пример**: зелёный круг нужно классифицировать как синий квадрат (класс 1) или как красный треугольник (класс 2). Если k = 3, то он классифицируется как 2-й класс, потому что внутри меньшего круга 2 треугольника и только 1 квадрат. Если k = 5, то он будет классифицирован как 1-й класс (3 квадрата против 2 треугольников внутри большего круга)

### Метрики регрессии.

Существует большое количество так называемых "метрик". Что это и как их вычислить?

 По сути своей "метрика" - это оценка качества реализованной модели. Метрику нужно выбирать с зависимости от задачи, которую мы пытаемся решить. Но метрики не единственный способ оценки качества.

Для оценки немаловажным показателем является "Loss", на русском потеря, то есть по сути то на сколько модель ошиблась в своём предсказании.

\begin{array}{|c|c|} \hline
 & y = 1 & y = 0 \\ \hline
\hat{y} = 1 & True Positive (TP) & False Positive (FP) \\ \hline
\hat{y} = 0 & False Negative (FN) & True Negative (TN) \\\hline
\end{array}

Эта таблица показывает различные исходы предсказаний для обьекта.

* **Аккуратнсть**

  $Accuracy=\frac{TP+TN}{TP+TN+FP+FN}$

* **Точность**

  $Precision=\frac{TP}{TP+FP}$

* **Полнота**

  $Recall=\frac{TP}{TP+FN}$

* **F-мера**

  $F\_score = \frac{2}{\frac{1}{Precision} + \frac{1}{Recall}}$

Реализуем метрики самостоятельно и сравним с метриками из sklearn

```python
import sklearn.metrics as metrics
```

```python
y_true = np.array([1, 0, 1, 0, 0])
y_pred = np.array([1, 1, 1, 0, 1])
```

```python
(y_true == 1) & (y_pred == 1)
```

```python
def accuracy(y_true, y_pred):
  return sum(y_true == y_pred) / len(y_true)

assert accuracy(y_test, y_predicted) == metrics.accuracy_score(y_test, y_predicted)
print('Accuracy score is {:.2f}'.format(metrics.accuracy_score(y_test, y_predicted)))
```

```python
def precision(y_true, y_pred):
  return np.sum((y_true == 1) & (y_pred == 1)) / np.sum(y_pred)

assert precision(y_test, y_predicted) == metrics.precision_score(y_test, y_predicted)
metrics.precision_score(y_test, y_predicted)
```

```python
def recall(y_true, y_pred):
  # your code here
  return np.sum((y_true == 1) & (y_pred == 1)) / np.sum(y_true)

assert recall(y_test, y_predicted) == metrics.recall_score(y_test, y_predicted)
metrics.recall_score(y_test, y_predicted)
```

```python
def f_score(y_true, y_pred):
  return 2 * precision(y_true, y_pred) * recall(y_true, y_pred) / (precision(y_true, y_pred) + recall(y_true, y_pred))

assert f_score(y_test, y_predicted) == metrics.f1_score(y_test, y_predicted)
print('F-score score is {:.2f}'.format(metrics.f1_score(y_test, y_predicted)))
```

Здесь я хочу показать возможности ML, как выпускник рекомендованных  мной специализаций Яндекс лицея по направлению ML и возможно дать ориентир для продвижения.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
from torchvision import transforms, datasets
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
# Загрузка и предобработка данных
transform = transforms.Compose([
    transforms.ToTensor()
])

train_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=128, shuffle=True)
```

```python
def training(num_epochs, model, device, latent_dim, optimizer, criterion, train_dataset, train_loader):
  for epoch in range(num_epochs):
      model.train()
      train_loss = 0.0
      for inputs, _ in train_loader:

          inputs = inputs.to(device)
          inputs = F.interpolate(inputs, size=(32, 32), mode='bilinear', align_corners=False)

          optimizer.zero_grad()
          outputs, mu, logvar = model(inputs)
          outputs = F.interpolate(outputs, size=(32, 32), mode='bilinear', align_corners=False)
          recon_loss = criterion(outputs, inputs)
          kl_divergence = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())

          loss = recon_loss + kl_divergence
          loss.backward()

          train_loss += loss.item()

          optimizer.step()

      print(f'Epoch {epoch+1}/{num_epochs}, Loss: {train_loss/len(train_dataset)}')
```

```python
# Определение архитектуры вариационного автоэнкодера
class VAE_1(nn.Module):
    def __init__(self, input_dim, latent_dim):
        super(VAE_1, self).__init__()

        self.encoder = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, 1, 1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.fc_mu = nn.Linear(64*8*8, latent_dim)
        self.fc_logvar = nn.Linear(64*8*8, latent_dim)

        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(latent_dim, 64, 4, 1, 0),
            nn.ReLU(),
            nn.ConvTranspose2d(64, 32, 4, 2, 1),
            nn.ReLU(),
            nn.ConvTranspose2d(32, 3, 4, 2, 1),
            nn.Sigmoid()
        )

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5*logvar)
        eps = torch.randn_like(std)
        return mu + eps*std

    def forward(self, x):
        x = self.encoder(x)
        x = x.view(-1, 64*8*8)
        mu = self.fc_mu(x)
        logvar = self.fc_logvar(x)
        z = self.reparameterize(mu, logvar)
        z = z.view(-1, latent_dim, 1, 1)
        x_recon = self.decoder(z)
        return x_recon, mu, logvar

# Обучение модели
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

latent_dim = 100
model_1 = VAE_1(input_dim=3, latent_dim=latent_dim).to(device)
optimizer = optim.Adam(model_1.parameters(), lr=0.001)
criterion = nn.BCELoss(reduction='sum')

num_epochs = 20
training(num_epochs, model_1, device, latent_dim, optimizer, criterion, train_dataset, train_loader)
```

```python
# Визуализация результата
model_1.eval()
with torch.no_grad():
    inputs, _ = next(iter(train_loader))
    inputs = inputs.to(device)
    inputs_upsampled = F.interpolate(inputs, size=(32, 32), mode='bilinear', align_corners=False)
    outputs, _, _ = model_1(inputs_upsampled)

    plt.figure(figsize=(10, 4))

    for i in range(5):
        plt.subplot(2, 5, i+1)
        plt.imshow(inputs[i].permute(1, 2, 0).cpu().numpy())
        plt.axis('off')

        plt.subplot(2, 5, i+6)
        plt.imshow(outputs[i].cpu().permute(1, 2, 0))
        plt.axis('off')

    plt.show()
```
